```python
import pandas as pd
import matplotlib.pyplot as plt
df_csv=pd.read_csv('LokSabha.csv')
df_csv
```

```python
plt.figure(figsize=(14, 8))
plt.bar(df_csv['States and Union Territories'], df_csv['No. Of Phases'], color='mediumseagreen')
plt.xticks(rotation=90)
plt.title('Number of Phases per State/UT')
plt.xlabel('States and Union Territories')
plt.ylabel('No. Of Phases')
plt.tight_layout()
plt.show()
```

```python
# Convert 'No. Of Phases' to numeric in case it's read as object/string
df_csv['No. Of Phases'] = pd.to_numeric(df_csv['No. Of Phases'], errors='coerce')

# Take top 10 for clarity
top10 = df_csv.sort_values('No. Of Phases', ascending=False).head(10)

plt.figure(figsize=(8, 8))
plt.pie(top10['No. Of Phases'], labels=top10['States and Union Territories'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 States/UTs by Number of Phases')
plt.axis('equal')
plt.show()
```
